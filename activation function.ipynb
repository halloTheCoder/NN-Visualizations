```python
%matplotlib inline 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# utility function
def make_plot(x, f, df, name):
    plt.figure(figsize = (12, 6))
    plt.title(name, fontsize = 20, fontweight = 'bold')
    plt.xlabel('z', fontsize = 15)
    plt.ylabel('Activation function value', fontsize = 15)
    sns.set_style("whitegrid")
    plt.plot(x, f, label = "f (z)")
    plt.plot(x, df, label = "f '(z)")
    plt.legend(loc = 4, prop = {'size': 15}, frameon=True, shadow=True, facecolor="white", edgecolor="black")
#     plt.savefig(name + ".png")
    plt.show()
```

### Sigmoid

$${\displaystyle f(x)={\frac {1}{1+e^{-x}}}}
\hspace{1cm}
{\displaystyle f'(x) = f(x)\cdot(1-f(x))}$$

```python
z = np.arange(-10, 10, 0.01)
f = 1/(1+ np.exp(-z))
df = f * (1 - f)
make_plot(z, f, df, 'Sigmoid')
```

### Tanh

$${\displaystyle f(x) = \operatorname {tgh} x={\frac {\sinh x}{\cosh x}}={\frac {e^{x}-e^{-x}}{e^{x}+e^{-x}}}}
\hspace{1cm}
{\displaystyle f'(x) = 1 - f(x)^{2}}$$

```python
z = np.arange(-10, 10, 0.01)
f = np.tanh(z)
df = 1 - f * f
make_plot(z, f, df, 'Tanh')
```

### ReLU

$${\displaystyle f(x) = \begin{cases}
0 & \text{ if } x<0 \\ 
x & \text{ if } x\geq 0 
\end{cases}}
\hspace{1cm}
{\displaystyle f'(x) = \begin{cases}
0 & \text{ if } x<0 \\ 
1 & \text{ if } x\geq 0 
\end{cases}}$$

```python
z = np.arange(-10, 10, 0.01)
f = z * (z > 0)
df = 1. * (z > 0)
make_plot(z, f, df, 'ReLU')
```

### Leaky ReLU

$${\displaystyle f(x) = \begin{cases}
0.01 \cdot x & \text{ if } x<0 \\ 
x & \text{ if } x\geq 0 
\end{cases}}
\hspace{1cm}
{\displaystyle f'(x) = \begin{cases}
0.01 & \text{ if } x<0 \\ 
1 & \text{ if } x\geq 0 
\end{cases}}$$

```python
z = np.arange(-10, 10, 0.01)
f = np.where(z > 0, z, 0.01 * z)
df = np.where(z > 0, 1, 0.01)
make_plot(z, f, df, 'Leaky ReLU')
```

### ELU

$${\displaystyle f(x) = \begin{cases}
\alpha \cdot (e^{x} - 1) & \text{ if } x<0 \\ 
x & \text{ if } x\geq 0 
\end{cases}}
\hspace{1cm}
{\displaystyle f'(x) = \begin{cases}
f(x) + \alpha & \text{ if } x<0 \\ 
1 & \text{ if } x\geq 0 
\end{cases}}$$

```python
z = np.arange(-10, 10, 0.01)
alpha = 2
f = np.where(z > 0, z, alpha * (np.exp(z) - 1))
df = np.where(z > 0, 1, f + alpha)
make_plot(z, f, df, 'ELU')
```

### SoftPlus

$${\displaystyle f(x)=ln(1+e^x)}
\hspace{1cm}
{\displaystyle f'(x) = \frac {1}{1+e^{-x}}}$$

```python
z = np.arange(-10, 10, 0.01)
f = np.log(1 + np.exp(z))
df = 1/(1 + np.exp(-z))
make_plot(z, f, df, 'SoftPlus')
```
